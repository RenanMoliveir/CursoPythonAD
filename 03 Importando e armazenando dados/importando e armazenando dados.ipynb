### Importando arquivos CSV

```python
import pandas as pd# importing the pandas library

#importing the data
arquivo = "dados/arquivo_csv_virgula.csv" #path to the file
#pd.read_csv(arquivo) #reading the file
```

```python
funcionario = pd.read_csv(arquivo)# reading the file
funcionario.head() #showing the first 5 lines of the file
```

```python
funcionario.dtypes #showing the data types of the columns
```

```python
# adding params to the read_csv function
funcionario = pd.read_csv(arquivo, sep= ',', parse_dates=['Data de Nascimento'], encoding="utf-8", thousands=".", decimal=",", dtype={"Idade":float})# reading the file with specific parameters
funcionario.head()
```

```python
funcionario.dtypes #showing the data types of the columns again to see the changes
```

```python
# importando o arquivo com o separador de ponto e vírgula
# importing the file with the semicolon separator
arquivo_pv = "dados/arquivo_csv_ponto_virgula.csv" #path to the file
funcionario_pv = pd.read_csv(arquivo_pv)# reading the file
funcionario_pv.head() 
```

```python
# Importando arquivo com pipe
# importing file with pipe
arquivo_txt = "dados/arquivo_csv_ponto_pipe.txt"#path to the file
funcionario_txt = pd.read_csv(arquivo_txt, sep='|')# reading the file with pipe separator
funcionario_txt.head() #showing the first 5 lines of the file with pipe separator
```

```python
# importing excel file
# importing archivo excel
arquivo_excel = "dados/vendas_2022.xlsx" #path to the file
vendas = pd.read_excel(arquivo_excel)# reading the excel file
vendas.head() #showing the first 5 lines of the excel file
```
